```python
# prompt: github graphqlを使ってteam listを取得したい

import requests
import os
from google.colab import userdata


# GitHub Personal Access Token (PAT)
#  - generate one at https://github.com/settings/tokens
#  - give it the "repo" scope at a minimum
GITHUB_TOKEN = userdata.get('GITHUB_TOKEN')

if not GITHUB_TOKEN:
  raise ValueError("GITHUB_TOKEN environment variable not set. Please set it to your GitHub Personal Access Token.")

headers = {
    "Authorization": f"Bearer {GITHUB_TOKEN}",
    "Accept": "application/vnd.github.v4+json"
}

def get_teams(org_name):
  query = """
  query($org: String!) {
    organization(login: $org) {
      teams(first: 100) { # Adjust 'first' if you expect more than 100 teams
        nodes {
          name
          slug
          description
          members {
            totalCount
          }
          repositories {
            totalCount
          }
        }
        pageInfo {
          endCursor
          hasNextPage
        }
      }
    }
  }
  """
  variables = {"org": org_name}

  response = requests.post("https://api.github.com/graphql", json={"query": query, "variables": variables}, headers=headers)

  if response.status_code == 200:
    data = response.json()
    teams = data["data"]["organization"]["teams"]["nodes"]
    return teams
  else:
    print(f"Error: {response.status_code} - {response.text}")
    return None

# Example usage: replace "your_organization_name" with the actual organization name
organization_name = "axelspace"
teams = get_teams(organization_name)

if teams:
  for team in teams:
    print(f"Team Name: {team['name']}")
    print(f"Team Slug: {team['slug']}")
    print(f"Description: {team['description']}")
    print(f"Members: {team['members']['totalCount']}")
    print(f"Repositories: {team['repositories']['totalCount']}")
    print("-" * 20)
```
